```python
## This is a simple ODE IVP probelm. Using the equation dy(t)/dt = t - y(t). We 
## are able to apply the scipy.integrate.solve_ivp function to evaluate with an 
## initial condition y(0) = 2. This is done over a specified time range 0 to 15. 
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

sol = solve_ivp(lambda t, y: t-y, [0, 15], [2], rtol = 1e-5)
plt.plot(sol.t, sol.y[0], 'k--s')
```

```python
## This reduces the order of an ODE by seperating y'' + y' + 2y = 0 into two 
## first order differential equations. 
import numpy as np
from scipy import integrate

def solvr(Y, t):
    return [Y[1], -2 * Y[0]-Y[1]]
a_t = np.arange(0, 25.0, 0.01)
asol = integrate.odeint(solvr, [1, 0], a_t)
print(asol)
```

```python
## Using the Euler method with smaller steps yields more accurate results. 
## These results are plotted against the actual value that the code should get to. 
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

f = lambda t, s: np.exp(-t)
h = 0.01
t = np.arange(0, 1 + h, h)
s0 = -1
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
